```python
def part1(n):
    # el input n debe ser N-2 donde N es la base que verdaderamente buscamos
    # en el resto de las funciones ocurrirá lo mismo
    l = []
    f_basis = basis(n)
    zeros_r = sy.Matrix.zeros(1,n)
    zeros_c = sy.Matrix.zeros(n+2,1)

    for e in f_basis:
        e=e.row_insert(n,zeros_r)
        e=e.row_insert(0,zeros_r)
        e=e.col_insert(n,zeros_c)
        e=e.col_insert(0,zeros_c)
        l.append(e)

    return l


def part2y3(n):
    l = []

    for r in range(1,n):
        e = sy.Matrix.zeros(n+2,n+2)
        e[0,1]=1; e[n+1,r+1]=1
        e[n+1,1]=-1; e[0,r+1]=-1
        l.extend([e,e.T])

    return l


def part4(n):
    l = []

    e = sy.Matrix.zeros(n+2,n+2)
    e[n+1,0]=1; e[n,n+1]=1; e[0,n]=1
    e[0,n+1]=-1; e[n,0]=-1; e[n+1,n]=-1
    l.append(e)

    e = sy.Matrix.zeros(n+2,n+2)
    e[n+1,n+1]=1; e[n,0]=1; e[0,1]=1
    e[0,0]=-1; e[n,n+1]=-1; e[n+1,1]=-1
    l.append(e)
    
    return l
    

def part5y6(n):
    l = []

    e = sy.Matrix.zeros(n+2,n+2)
    e[0,1]=1; e[math.ceil(n/2),2]=1
    e[0,2]=-1; e[math.ceil(n/2),1]=-1
    l.extend([e,e.T])

    e = sy.Matrix.zeros(n+2,n+2)
    e[0,n-1]=1; e[math.ceil(n/2),n]=1
    e[0,n]=-1; e[math.ceil(n/2),n-1]=-1
    l.extend([e,e.T])

    for r in range(1,n-2):
        e = sy.Matrix.zeros(n+2,n+2)
        e[0,r+1]=1; e[1,r+2]=1
        e[0,r+2]=-1; e[1,r+1]=-1
        l.extend([e,e.T])
    
    return l


def part7(n):
    l = []

    e = sy.Matrix.zeros(n+2,n+2)
    e[n+1,0]=1; e[n,1]=1; e[1,n-1]=1; e[2,n]=1
    e[n,0]=-1; e[n+1,1]=-1; e[1,n]=-1; e[2,n-1]=-1
    l.append(e)

    e = sy.Matrix.zeros(n+2,n+2)
    e[0,0]=1; e[1,1]=1; e[n,n-1]=1; e[n-1,n]=1
    e[1,0]=-1; e[0,1]=-1; e[n,n]=-1; e[n-1,n-1]=-1
    l.append(e)

    return l



```

```python
def basis(N):
    """obtener una base para los n-átomos
     por cómo esta construido, en los otros métodos (las partes) hay que introducir
     de input n-2"""
    
    if N==3:
        return base3
    elif N==4:
        return base4
    elif N==5:
        return base5
    elif N==6:
        return base6
    else:
        return part1(N-2)+part2y3(N-2)+part4(N-2)+part5y6(N-2)+part7(N-2)
```

```python
# Por si el lector no se fía, esta función comprueba si la base que se obtiene en basis(n) efectiva
# mente proporciona una lista de cuadrados mágicos que sean linealmente independientes (que formen una base)
 
def check(n):
    c = basis(n)

    w = 0

    #si son cuadrados mágicos
    ones = sy.Matrix.ones(n,n)
    zeros = sy.Matrix.zeros(n,n)
    for a in c:
        w+=1
        if ones*a!=zeros or a*ones!=zeros:
            print('WAIT 1',w)
            sy.pprint(a)
        diag1 = 0
        diag2 = 0
        for i in range(n):
            diag1 += a[i,i]
            diag2 += a[n-1-i,i]
        if diag1!=0 or diag2!=0:
            print('WAIT 2',w)
            sy.pprint(a)
    
    #si son linealmente independientes
    M = sy.Matrix.zeros(1,n**2)
    for a in c:
        M=M.row_insert(1,a.reshape(1,n**2))
    M.row_del(0)
    if M.rank()!=len(c):
        print('WAIT 3',M.rank())
    
```

```python
def random_ms(n,m,format='real', interval=100):
    """
    Devuelve m cuadrados mágicos aleatorios de orden n
    Si format='real', los valores del CM son reales
    Si format='entero', los valores del CM son enteros
    Cuanto más alto sea interval, mayores tenderán a ser los valores de los CMs
    """
    L = []
    b = basis(n)

    if format=='real':
        for i in range(m):
            p = sy.Matrix.zeros(n,n)
            for e in b:
                p += interval*random.random()*e
            L.append(p)

    if format=='entero':
        for i in range(m):
            p = sy.Matrix.zeros(n,n)
            for e in b:
                p += random.randint(-interval,interval)*e
            L.append(p)

    return L
```
